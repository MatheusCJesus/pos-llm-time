### Concatenar todos os files

```python
files = os.listdir("../HIST_PAINEL_COVIDBR_21out2023/")
df_all = pd.DataFrame()
for file in files:

    tmp = pd.read_csv(f"../HIST_PAINEL_COVIDBR_21out2023/{file}", sep=";")
    df_all = pd.concat([df_all, tmp], axis="rows")
```

```python
df_all_sp = df_all[df_all["estado"] == "SP"]
df_all_sp
```
